# Quiet checkers graph API demo

This notebook demonstrates the library API on a bounded run. The graph stores distinct packed states once and keeps quiet transitions in round-layered successor maps. DataFrames use stable indexes such as `state_index` and `transition_index` as handles for display/path methods.

```python
%matplotlib inline

from pathlib import Path
import sys
from importlib import reload

repo_root = Path.cwd()
if not (repo_root / "src").exists():
    repo_root = repo_root.parents[1]

dev_notebook_dir = repo_root / "notebooks" / "dev"
if str(dev_notebook_dir) not in sys.path:
    sys.path.insert(0, str(dev_notebook_dir))

import quiet_notebook_live
quiet_notebook_live = reload(quiet_notebook_live)

from pycheckers.quiet import QuietPositionGraph, run_quiet_graph
from quiet_notebook_live import run_live_quiet_exploration
```

## Bounded graph run

```python
graph = run_quiet_graph(max_rounds=4)
graph.summary()
```

```python
rounds_df = graph.rounds_df()
states_df = graph.states_df()
transitions_df = graph.transitions_df()
boundary_df = graph.boundary_states_df()

rounds_df
```

## Indexed transitions

Use `predecessor_index` or `next_index` to inspect edges without expanding a tree.

```python
predecessor_index = 0
graph.transitions_df(
    predecessor_index=predecessor_index,
    include_state_fields=True,
)
```

```python
shared_next_index = int(transitions_df["next_index"].value_counts().loc[lambda s: s > 1].index[0])
graph.predecessors_df(shared_next_index, include_state_fields=True)
```

## Path-count DataFrames

```python
path_count_distribution_df = graph.path_count_distribution_df(max_path_count=5)
least_path_count_df = graph.least_path_count_states_df(limit=20)
unique_path_df = graph.unique_path_states_df(limit=20)

path_count_distribution_df
```
