# pycheckers primitive rule catalog

A visual check of the compact `Ruleset` runtime rules for black and white moves. Each displayed output row shows one runtime rule with a condition board, an effect board, and a compact rule summary. The `king`, `capture`, and `promotion` bits describe how the rule should be interpreted without duplicating that information as board masks. Multi-jump turns are intentionally excluded here; jump chaining is handled by turn generation.


```python
import pandas as pd

from pycheckers import Ruleset

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 240)
```

```python
ruleset = Ruleset.american()
runtime_rules = ruleset.to_dataframe()
print(f"primitive runtime rule rows: {len(runtime_rules)}")
runtime_rules
```

```python
_ = ruleset.plot(
    size=3.2,
)
```
